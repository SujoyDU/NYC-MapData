```python
import pandas as pd
import numpy as np
```

```python
restaurants = pd.read_csv('DOHMH_NYC_RestaurantInspections.csv')
restaurants.head()
```

```python
restaurants.shape
```

```python
# sort_values by 
x = restaurants.sort_values(["CAMIS", "INSPECTION DATE"])
x.head(20)
```

```python
x.shape
```

```python
y = x.sort_values('INSPECTION DATE').drop_duplicates('CAMIS',keep='last')
y.head()
```

```python
y.shape
```

```python
z = y.sort_values(["CAMIS", "INSPECTION DATE"])
z.head()
```

```python
z.to_csv('NYC_RestaurantInspection_deduped.csv', index = False)
```
